```python
x=k1.drop('Species',axis=1)
y=k1['Species']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
prediction
```

```python
print(accuracy_score(y_test,prediction))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm= confusion_matrix(y_test,prediction)
cm
```

```python
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
result = confusion_matrix(y_test,prediction)
print("Confusion Matrix:")
print(result)
result1 = classification_report(y_test,prediction)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(y_test,prediction)
print("Accuracy:",result2)
```
